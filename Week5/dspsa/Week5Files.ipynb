### Some practice on patterns and regex
- find all numbers in a string (not individual digits)
- all numbers surrounded by alphabetical chars (upper and lower)
- all strings containing only consonants --> try to use ^
- all the strings that exist out of the letters of "lorem" --> it can only exist out of letters from lorem
- all strings that contain the letters of a given word AND are of equal length --> tip: rf""

```python
import re

text = """Lorem ipsum dolor sit amet, consectetur adipiscing elit. Nullam vehicula pulvinar leo, vehicula 
commodo Anthony augue consequat a. Cras finibus varius dapibus. Donec ynothna eu risus ut eros lacinia luctus. Ut sit amet 
euismod sem. Pellentesque ornare commodo mauris dictum ornare. Suspendisse molestie metus sit amet mi consectetur, 
ac tristique nnatoyh urna laoreet. Mauris eget congue odio. Nulla non odio aliquam, tempus dolor sit amet, aliquam risus. 
Morbi fringilla tincidunt gravida.
"""
#- find all numbers in a string (not individual digits)
regex = r"\b\d{2,}\b"
print(re.findall(regex, text.lower()))

#- all numbers surrounded by alphabetical chars (upper and lower)
regex = r"\b[A-z]+(\d+)[A-z]+\b"
print(re.findall(regex, text.lower()))

#- all strings containing only consonants --> try to use ^
regex = r"\b[^aeiou\W\s\d]+\b"
print(re.findall(regex, text.lower()))

#- all the strings that exist out of the letters of "lorem" --> it can only exist out of letters from lorem
regex = r"\b[loremLOREM]+\b"
print(re.findall(regex, text.lower()))

#- all strings that contain the letters of a given word AND are of equal length --> tip: rf""
letters = input()
regex = rf"\b[{letters}]{{{len(letters)}}}\b"
print(re.findall(regex, text.lower()))
```

### Last exercise on regex
write a python function --> validate_email() that takes in an email address and returns a boolean that tells the
user whether the email is valid or not.

- Emailaddress has to contain a @
- First part of the email (before @) can only include letters, digits, hyphens ( - ), underscores ( _ ) and dots. Has to be at
least 1 char.
- part after the @ --> has to be at least 2 chars, and only letters, dots . and hyphens ( - )
- cannot end in a dot.

```python
import re

def validate_email(email):
    if "@" not in email:
        return False
    
    parts = email.split("@")
    if len(parts) != 2:
        return False
    
    regex = r"^[\w\.\-_]+$"
    if not re.match(regex, parts[0]):
        return False
    
    regex = r"^[A-z\.\-]{2,}$"
    if re.match(regex, parts[1]) is None:
        return False
    
    if email[-1] == ".":
        return False

    return True

print(validate_email("anthony.coppens@thomasmore.be"))
print(validate_email("x!,anthony.coppensx!:;@thomasmore.be"))
```
