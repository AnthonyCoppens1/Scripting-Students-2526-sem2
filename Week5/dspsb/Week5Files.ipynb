### A last exercise on regex
Checking emails

Write a python function called --> validate_email
--> takes an email as input and returns a boolean indicating whether the email is valid or not.

- email must contain ONLY 1 @ --> separation??
- part before "@" --> at least 1 char and only letters, numbers, dots, hyphens ( - ) and underscores ( _ )
- part after "@" --> at least 2 chars and only letters, dots and hyphen ( - )
- cannot end with a dot .

```python
import re

def validate_email(email):
    if "@" not in email:
        return False
    
    parts = email.split("@")
    if len(parts) != 2:
        return False
    
    regex = r"^[\w\.\-_]+$"
    if re.match(regex, parts[0]) is None:
        return False
    
    regex = r"^[A-z\.\-]{2,}$"
    if re.match(regex, parts[1]) is None:
        return False
    
    if email[-1] == ".":
        return False

    return True

print(validate_email(r"anthony.coppens@thomasmore.be"))
print(validate_email(r"scammer20000$x3!@scamvil@loser.net"))
```

## Walking through a directory

```python
import os
from pathlib import Path

#os.chdir(Path.home()) --> changes directory to user profile
for folderName, subfolders, filenames in os.walk(Path.cwd()):
    print(f"current folder is: {folderName}")

    for subfolder in subfolders:
        print(f"SUBFOLDER of {folderName} --> {subfolder}")
    
    for filename in filenames:
        print(f"FILE INSIDE {folderName} --> {filename}")
    
    print("-------------")
```

```python
#printing and searching for a file in a certain path
from pathlib import Path
path_object = Path.cwd()

for file in path_object.glob("*.txt"):
    print(file)

for file in path_object.glob("*[0-9]*"):
    print(f"file with number inside: {file}")
```
