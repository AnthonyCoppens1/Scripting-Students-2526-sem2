## JSON vs XML
JSON is a text based file format mainly usd for storing and transmitting data, usually this is typically done betweena webserver and a client --> do keep in mind, this is normally the standard for data transmission ATM

XML: more complex, requires a setup of a tag structure and is worse to process in a lot of coding languages

https://codebeautify.org/jsonviewer

```python
import json

#string format of json code --> examples
json_data = '{"name": "Anthony", "age" : 30, "study": "computer science"}'
data = json.loads(json_data) #string load

print(data["name"]) #keep in mind you're cycling through a dictionary --> result is a dictionary or a part of it.


#file load (option 2)
json_file = open("books.json", "r")
data = json.loads(json_file.read())
print()

print(f'First book title: {data["books"][0]["title"]}')
print(data)

for book in data["books"]:
    print(book["title"])

json_file.close()
```

```python
import json, requests

url = "https://api.kanye.rest/"
response = requests.get(url)

#print(response)
#print(response.status_code)

text = response.text
data = json.loads(text)

print(data)
print(f'''\nKanye's daily quote: {data["quote"]}''')
```

```python
#BUILDING OUR OWN "PROFESSIONAL" LOVE CALCULATOR, PROVEN BY THE ONE AND ONLY ONDREY

import requests

url = "https://the-love-calculator.p.rapidapi.com/love-calculator"
firstname = input()
secondname = input()
querystring = {"fname": firstname,"sname": secondname}

headers = {
	"x-rapidapi-key": "fd6f435c07msh3de62319fcfa275p143e5cjsn4293d86a9f09",
	"x-rapidapi-host": "the-love-calculator.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.get(url, headers=headers, params=querystring)
data = response.json()

print("DEAR LOVE ENJOYERS\n")
print(f"BEHOLD THE LOVE % BETWEEN THE FOLLOWING 2 PEOPLE: {firstname} and {secondname}")
print(f'Their love %: {data["percentage match: "]}')
print(f'Message from the company: {data["result: "]}')
```

```python
#TRANSLATING WHATEVER YOU LIKE
import requests

url = "https://free-google-translator.p.rapidapi.com/external-api/free-google-translator"

querystring = {"from":"en","to":"nl","query":"Learning Dutch is very easy, that's what Joseph said."}

payload = { "translate": "rapidapi" }
headers = {
	"x-rapidapi-key": "fd6f435c07msh3de62319fcfa275p143e5cjsn4293d86a9f09",
	"x-rapidapi-host": "free-google-translator.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers, params=querystring)

data = response.json()
print(data["translation"])
```
