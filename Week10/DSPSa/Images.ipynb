### Creating images

```python
from PIL import Image

dogface = Image.open("dogface.jpg")
img = Image.open("dog.jpg")

locationToPasteUpon = img.copy()
locationToPasteUpon.paste(dogface, (0,0))
locationToPasteUpon.paste(dogface, (img.width - dogface.width, 0))
locationToPasteUpon.show()
locationToPasteUpon.save("dogdogdog.jpg")
```

```python
from PIL import Image
img = Image.new("RGBA", (100,200), "white")

dogface = Image.open("dogface.jpg")
img = Image.new("RGBA", (dogface.width * 10, dogface.height * 10), "white")
copy = dogface.copy()

for i in range(0,10):
    for j in range(0,10):
        if (i+j) % 2 == 0 :
            img.paste(copy, (i * dogface.width, j * dogface.height))
        else:
            rotated = dogface.rotate(180)
            img.paste(rotated, (i * dogface.width, j * dogface.height))

img.show()
```

### Transforming images

```python
from PIL import Image
img = Image.open("dog.jpg")
img.rotate(90).show()
img.rotate(180).show()
img.rotate(-90).show()

img.transpose(Image.FLIP_LEFT_RIGHT).show()
img.transpose(Image.FLIP_TOP_BOTTOM).show()

img.resize((int(img.width / 2), img.height // 2)).show()
flat = img.resize((int(img.width * 7), img.height // 2))
flat.save("flatdog.jpg")
```

### looping over dog pixel by pixel


```python
from PIL import Image
img = Image.open("dog.jpg")
for x in range(0, img.width):
    for y in range(0, img.height):
        r,g,b = img.getpixel((x,y))
        inverted = (255-r, 255-g, 255-b)
        img.putpixel((x,y), inverted)

img.show()
```

### finding contours

```python
import cv2

img = cv2.imread("dog.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#threshold and a retina for the binary image
ret, thresh = cv2.threshold(gray, 150, 255, 0)

#finding contours
contours, hierarchy = cv2. findContours(thresh, cv2.RETR_TREE, cv2. CHAIN_APPROX_SIMPLE)

#draw the lines
contour_img = img.copy()
cv2.drawContours(contour_img,contours, -1, (255,0,0), 2)
cv2.imwrite("dogContour.jpg", contour_img)
```

## Object detection
### Classical approach
Haar cascade classifier

```python
import cv2
img = cv2.imread("madrid.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x + w, y + h), (255,0,0), 2)

cv2.imwrite("MadridFaces.jpg", img)
```
