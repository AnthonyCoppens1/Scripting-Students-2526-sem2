### chessboard of birdies
Why? Because we love them!

```python
from PIL import Image
lilBirdie = Image.open("lilBirdie.jpg")
copy = lilBirdie.copy()
flippedBirdie = lilBirdie.rotate(180).copy()
img = Image.new('RGBA', (lilBirdie.width*10, lilBirdie.height*10), "white")
for i in range(0,10):
    for j in range(0,10):
        if (i+j) % 2 == 0:
            img.paste(copy, (i * lilBirdie.width, j * lilBirdie.height))
        else:
            img.paste(flippedBirdie, (i * lilBirdie.width, j * lilBirdie.height))

img.show()
```

### Looping over and inverting the bird

```python
from PIL import Image
img = Image.open("bird.jpg")
for x in range(0, img.width):
    for y in range(0, img.height):
        r,g,b = img.getpixel((x,y))
        inverted = (255-r, 255-g, 255-b)
        img.putpixel((x,y), inverted)

img.save("inverted.jpg")
img.show()
```

### Finding contours
useful for shape analysis, measurements, finding boundaries, ...

```python
import cv2

#load and to convert to grayscale
img = cv2.imread("bird.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#threshold for binary image
ret, thresh = cv2.threshold(gray, 127, 255, 0)

#finding our contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 0, 255), 2)
cv2.imwrite("birdWithContours.jpg", contour_img)
```

## Object detection
### classical approach

```python
import cv2
img = cv2.imread("thailand2.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
face = face_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5)

for (x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x + w, y + h), (255,0,0), 2)

cv2.imwrite("thailandFaces.jpg", img)
```
