## Creating arrays
array of elements has to be of the same datatype

```python
import numpy as np

list = [1, 5, 8, 10, 3]
print(list)

array = np.array(list)
print(array)

print(list * 2) #list is seen as a whole, it will duplicate the list and paste itself behind the first occurence.
print(array * 2) #treats every element separately to multiply

list.append(2)
#array.append(2)
#array.add(2) --> none of these 2 work, because arrays have a fixed size!
```

```python
import numpy as np

#1D arrays
print(np.arange(10)) #--> fill an array with x numbers from 0 till x
print(np.zeros(10)) #fill an array with x 0's
print(np.ones(10)) 
#print(np.twos(10)) --> does not work, only ones and zeros

print(np.full(10, 5)) #fill array with certain length with specific number

print(np.random.randint(-5, 6, 10)) #--> remove the last digit and get only 1 number
print(np.random.random(10)) #x amount of random floats between 0 and 1.
```

## Reshaping & converting an array

```python
import numpy as np
array = np.random.randint(-5, 6, 10)

print(array)
print(len(array)) #works but only in 1D
print(array.shape)
print(array.ndim)

#checking datatype
print(f"array is of datatype: {array.dtype}")

#conversions between types of arrays
arrayFloater = array.astype(float) #str int float ...
print(arrayFloater)

arrayString = array.astype("S")
print(arrayString)

# i - integer / b - boolean / f - float / M - datetime / S - string / but wording also works (float, str, int, ...)


#2D array
array2D = np.reshape(array, (2,5)) #first number = rows, second number = columns
print(array2D)

array2D = np.reshape(array, (5,2))
print(array2D)

#array2D = np.reshape(array, (3,3))
#array2D = np.reshape(array, (4,3))
#print(array2D)

#has to perfectly fit!
```

## Slicing arrays

[start : end : step]
- start: number (index), included
- end: number (index), NOT included
- step: the size of a jump between values

```python
import numpy as np

array = np.array([6, 8, 2, 9, 0, 1, 3, 5, 7, 4])
print(array)

#I want the first 4 numbers of the array
print(array[0:4:1]) #[6 8 2 9]
print(array[:4:1])
print(array[:4:])

#I want the 8 2 9 0
print(array[1:5:])
print(array[1:4:2])

print(array[::])
#reverse
print(array[::-1])

#slice the last 4, but don't use the positive index
print(array[-4::])

print(array[-1:0:-1])
print(array[0:10:-1]) #prints nothing
```

### Exercise on slicing
we want to generate an array of the temperature measurements EVERY hour throughout a day. (24 measurements)
generate random temps: between 0 and 25 (use random ints)
--> use slicing to isolate periods and analyse them
- earlyBirdFeedsTheWorm: all temps first 6 hours of the day (includes the temp at 00:00 --> 06:00)
- everyoneAwakeBEFORENOON: 09:00 - 13:00 (again the temps here)
- siesta: 14:00 - 16:00
- lateNightOwls: because they live in a different time, reverse the slice (LAST 4 MEASUREMENTS)
- trending: every 5 hours, startin at 00:00, 05:00, 10:00...

```python
import numpy as np
import random

#temps = [random.randint(0, 26) for x in range(24)]
#measurements = np.array(temps)

measurements = np.random.randint(0, 26, 24)
print(measurements)

earlyBirdFeedsTheWorm = measurements[:6:]
print(earlyBirdFeedsTheWorm)

everyoneAwakeBeforeNOON = measurements[9:13:]
print(everyoneAwakeBeforeNOON)

siesta = measurements[14:17:]
print(siesta)

lateNightOwls = measurements[-1:-5:-1]
print("LAST 4", lateNightOwls)
lateNightOwls = measurements[:-4]
print(lateNightOwls)


trending = measurements[::5]
print(trending)
```

## Working with arrays

```python
import numpy as np

a = np.arange(0,10)
b = np.arange(9,-1,-1)
print(a, b)

print(a + b)
print(a - b)
print(a * b)

print(np.where(a % 2 == 0, a + 5, a))
print(np.where(a % 2 == 0))

array = np.where(a % 2 == 0)
print(array[0]) #the regular np.where with only a condition and no True or False, returns a tuple with the indexes
#of the values where the condition is met.

print(np.where(b>5))
meetingCondition = np.where(b>5)[0]
print(meetingCondition)

b = np.reshape(b, (2,5))
print(b)
print()

print(b.sum(axis = 0)) #sum of all columns each counted up together
print(b.sum(axis = 1)) #sum of each row
print(b.cumsum(axis = 1))

print()
print(a.sum())
print(a.min())
print(a.max())

print(np.sqrt(a)) #square root of each number in a
print(np.sqrt(a.max()))
```
