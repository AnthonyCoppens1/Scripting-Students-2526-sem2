## Strings
### Raw strings

```python
s = "This is a regular string"
print(s)

s = 'This is also a string, but with single quotes'
print(s)

s = "This is a sentence with a single quote ' inside of it"
print(s)

s = "This is a sentence with a single ' and a double \" quote"
print(s) #escaping the middle double quote to truly show the symbol

s = "A sentence with single ', double \" and newline \n at the very back"
print(s) #\n is going to put the last half on a new line

s = r'''A sentence with single ', double " and newline \n at the very back'''
print(s) #raw --> r in front, to prevent any meaning of parts

#can use triple double quotes or triple single quote to go beyond regular limits

s = """This is a sentence with a single ' and a double " quote"""
print(s)
```

### Slicing a string
[ start : end : step ]
<br>end goes one beyond the maximum index you need, end is not included!

```python
s = "Anthony's course is sometimes amazing, sometimes boring"

#grab "course"
print(s[10:16:])

#reverse?
print(s[::-1])

#grab "boring"
print(s[-6::])
print(s[-len("boring")::]) #if you're too lazy or it's too difficult to count the indexes

print()
print("sometimes" in s)
print("Sometimes" in s)
print("Sometimes" not in s)
```

### Splitting and joining strings
splitting results in a list or strings in python, joining is the same as grouping it all together again

```python
s = "I love Python scripting!"
list = s.split() #default split is on space
print(list) #pay attention for things like !.?+:"'

list = s.split("Python")
print(list)

s = "1, 2, 3, 4, 5, 7, 9, 10"
list = s.split(", ")
print(list)

listOfIntegers = [int(x) for x in list]
print(listOfIntegers)

#for x in list:
#    listOfIntegers.append(int(x))

#joining strings
print(".".join(list))
#print("_".join(listOfIntegers)) --> impossible, cannot join strings and integers together, previous example
#was all only strings
```

### Usefull methods

```python
s = "This is a STRING with numbers 7 8 1 and everything you can think of"
# if it's uppercase --> lowercase / lowercase --> uppercase / digit --> "NUMBER" / space --> "."
#can use either isnumeric or isdigit, but beware of their usecases
for character in s:
    if character.isupper():
        print(character.lower(), end="")
    elif character.islower():
        print(character.upper(), end="")
    elif character.isdigit():
        print("NUMBER", end="")
    elif character.isspace():
        print(".", end="")
```

### Stripping a string
removing spaces up front and in the back

```python
s = "             a lot of spaces             "
print(s)
print("*", s, "*")
print("*", s.rstrip(), "*")
print("*", s.lstrip(), "*")
print("*", s.strip(), "*")

#small extra detail (not related to stripping)
character = "a"
print(ord(character))
i = 98
print(chr(i))
```

### Exercise: PigLatin
- word starts with a consonant --> add the consonant to the back of the word and add "ay"
--> e.g. bed --> edbay
- word starts with a double consonant --> add both consonants to the back and add "ay"
--> e.g. child --> ildchay
- word starting with a vowel --> don't change anything, but add "way" to the back
--> e.g. Anthony --> Anthonyway

Hello is Anthony and I am here today to teach you some Python
ellohay isway Anthonyway andway Iway amWay erehay odaytay otay eachtay ouyay omesay thonPyay

Write a function that can transfort word by word from a given sentence

```python
def PigLatin(word):
    if word[0].lower() in "aeiou":
        return word + "way"
    elif word[1].lower() not in "aeiou":
        return word[2::] + word[0:2:] + "ay"
    else:
        return word[1::] + word[0] + "ay"
    
sentence = input()
print(sentence)
for word in sentence.split():
    print(PigLatin(word), end=" ")
```

```python
#if you are uncertain of how a library works -->
help("numpy")
```

```python
import re

text = """Lorem ipsum dolor sit amet, consectetur adipiscing elit. Donec dapibus ex 5678 egestas ultrices placerat. 
Fusce posuere vehicula justo. 1234 Donec egestas dignissim erat, eu sodales odio malesuada sed. Phasellus semper odio 
ex, ac 1925 finibus lacus ornare scele7293risque. Class aptent taciti sociosqu ad litora torquent per conubia nostra, per 
inceptos himenaeos. Pellentesque habitant morbi tristique senectus et netus et malesuada fames ac turpis egestas. 
Donec et 6981 pretium lorem, et tincidunt leo."""

regex = r"[A-Z][a-z]+"
list = re.findall(regex, text)
print(list)

regex = r"\b\w{4}\b"
list = re.findall(regex, text)
print(list)

regex = r"\b\d{4}\b"
list = re.findall(regex, text)
print(list)

regex = r"Donec"
list = re.findall(regex, text)
print(list)

print(re.search(regex, text))
result = re.search(regex, text)
print(result.group()) #first match --> showing me the given match result
print(result.span()) #shows me the indexes of the contained match

regex_object = re.compile(regex) #it's like looking for donald trump in the epstein files
print(regex_object.match(text))

print(regex_object.split(regex))
```
