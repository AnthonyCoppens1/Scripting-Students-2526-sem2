### Splitting and joining
--> splitting a string into a list of strings / getting the list and returning a string

```python
s = "I love Python scripting!"
list = s.split()
print(list) #BEWARE OF END OF SENTENCE CHARS

list = s.split("n")
print(list)

list = s.split("Python")
print(list)

s = "1, 2, 3, 4, 5, 6, 8, 10"
list = s.split()
print(list)

list = s.split(", ") 
print(list)

#can also ignore spaces whenever converting.
listOfINTS = [int(x) for x in list]
print(listOfINTS)

print(".".join(list))
#print("_".join(listOfINTS)) --> impossible to join a list of integers with a given string
```

### Useful methods

```python
s = "This is a STRING with numbers 7 8 9"

for c in s: #for every character in s
    if c.isupper():
        print(c.lower(), end="")
    elif c.islower():
        print(c.upper(), end="")
    elif c.isnumeric():
        print("NR", end="")
    elif c.isspace():
        print(".", end="")
```

```python
s = "         a lot of spaces           "
print(s)
print("*", s, "*")
print("*", s.rstrip(), "*")
print("*", s.lstrip(), "*")
print("*", s.strip(), "*")
```

### Exercise on piglatin
- word starts with a consonant --> add the consonant to the back and add "ay"
--> happy --> appyhay
- word starts with 2 consonants --> add both to the back and add "ay"
--> child --> ildchay
- word starting with a vowel --> keep the word and add "way"
--> awesome --> awesomeway

Anthony teaches scripting
Anthonyway eachestay riptingscay

```python
def PigLatin(word):
    if word[0].lower() in "aeiou":
        return word + "way"
    else:
        if word[1].lower() not in "aeiou":
            return word[2::] + word[0:1:] + "ay"
        else:
            return word[1::] + word[0] + "ay"

sentence = input()
for word in sentence.split():
    print(PigLatin(word), end=" ")

#this is a test to see what
```
