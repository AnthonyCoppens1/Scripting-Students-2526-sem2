```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meteoblue.com/nl/weer/webcams/madrid_spanje_3117735"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

images = soup.find_all("a", {'class' : 'webcam-img'})
pictureLinks = []
for tag in images:
    pictureLinks.append(tag.img["src"])

index = 1
for newURL in pictureLinks:
    filename = "Webcams/original" + str(index) + ".jpg"
    r = requests.get(newURL, allow_redirects=True)
    with open(filename, "wb") as file:
        file.write(r.content)
    index += 1
```

### Reformatting dates and times

```python
import datetime as dt
import time

print(time.ctime())
print(dt.datetime.now())

date = dt.datetime(2000, 2, 10, 0, 0, 0)
print(date)
print(date.year + 1)

#date from a number of seconds
print(dt.datetime.fromtimestamp(1770000000))
print(dt.datetime.fromtimestamp(time.time()).year + 1)

#FORMATTING --> strftime()
print()
print(date)
print(date.strftime("%d %B %y"))
print(date.strftime("%D"))

date += dt.timedelta(days=1000)
print(date.strftime("%d %b %Y"))
```

## Multithreading
Python default program always runs in single threaded manner, so it only executes one thing at a time.
A multithreaded program can execute multiple things at the same time.

```python
import time

def sleep():
    time.sleep(5)
    print("HELLOOOOOOO I AM AWAKE")

def multi(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')

#SINGLE THREADED
print("START")
sleep()
multi(8)
print("STOP")
```

```python
import time, threading

def sleep():
    time.sleep(5)
    print("HELLOOOOOOO I AM AWAKE")

def multi(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')


#MULTITHREADED WAY
sleeper = threading.Thread(target=sleep)
multiplicator = threading.Thread(target=multi(8))

print("START")
sleeper.start()
multiplicator.start()
print("STOP")
```

```python
import time, threading

def function1():
    for i in range(1, 11):
        print(i, end=" ")
        time.sleep(0.5)

def function2():
    for c in "ABCDEFGHIJ":
        print(c, end=" ")
        time.sleep(0.5)
    
thread1 = threading.Thread(target=function1)
thread2 = threading.Thread(target=function2)

print("START")
thread1.start()
thread2.start()
thread1.join()
thread2.join()
print("STOP")

#if you encounter CONCURRENCY, add 0.01 seconds to one of the timers.
```
