## Big recap on webscraping
scraping all images from meteoblue.com

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meteoblue.com/nl/weer/webcams/chicago_verenigde-staten_4887398"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
imgTags = soup.find_all("a", {'class' : "webcam-img"})
links = []
names = []
for tag in imgTags:
    links.append(tag.img["src"])
    names.append(tag.img["src"].split("/")[-2] + ".jpg")

for index in range(len(links)):
    filename = "Meteoblue/" + names[index]
    newResponse = requests.get(links[index], allow_redirects=True)
    with open(filename, 'wb') as file:
        file.write(newResponse.content)
```

```python
# scrape the lineup from LETITROLL and do something with it
import requests
from bs4 import BeautifulSoup

url = "https://letitroll.eu/event/let-it-roll-2026/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

outerLayer = soup.find_all("div", {'data-id' : "07fbfde"})
names = outerLayer[0].get_text(strip=True).split("|")
allNames = []

for index in range(len(names)):
    if "B2B" in names[index]:
        allNames.append(names[index].split("B2B"))
    else:
        allNames.append(names[index])

ENDOFALLNAMES = []
for combo in allNames:
    if combo.__class__ == list:
        for name in combo:
            ENDOFALLNAMES.append(name.strip())
    else:
        ENDOFALLNAMES.append(combo)

print(ENDOFALLNAMES)
```

## Multithreading and difference between single and multi
Keep in mind python always runs default single threaded, meaning: 1 process at a time and whenever it is 
finished, you start the next.
With multithreading we can start multiple processes at the exact same time and run them together

```python
def sleep():
    time.sleep(5)
    print("HELLOOOOOO I AM AWAKE")

def multiplication(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')

#SINGLE THREADED
print("START")
sleep()
multiplication(8)
print("STOP")
```

```python
import threading
def sleep():
    time.sleep(5)
    print("HELLOOOOOO I AM AWAKE")

def multiplication(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')


sleeperThread = threading.Thread(target=sleep)
muliplicationThread = threading.Thread(target=multiplication(8))

print("START")
sleeperThread.start()
muliplicationThread.start()
print("STOP")
```

### Exericse on threading
create 2 functions that print something at the exact same time
- function 1 --> print 1 through 10 every half a second a number
- function 2 --> print a - j every half a second a letter
wait for both and then finish

```python
import time, threading
def function1():
    for i in range(1,11):
        print(i, end=" ")
        time.sleep(0.5)

def function2():
    for c in "ABCDEFGHIJ":
        print(c, end=" ")
        time.sleep(0.51)

thread1 = threading.Thread(target=function1)
thread2 = threading.Thread(target=function2)

print("START")
thread1.start()
thread2.start()
thread1.join()
thread2.join()
print("STOP")
```
