## Input and output
Conversion between datatypes
- int --> whole numbers
- float --> decimal value
- str --> string of text

```python
# find the sign of x
x = input()
print(x)
print(type(x))

int_x = int(x)
print(int_x)
print(type(int_x))

float_x = float(x)
print(float_x)
print(type(float_x))

backToString = str(float_x)
print(backToString)
print(type(backToString))
```

## Combining outputs --> concatenation
- '+' --> combining 2 strings or adding 2 values together if they can be calculated with.
- ',' --> split between 2 datatypes to be able to combine them into one piece of output/text
- f"" --> allows us to add variables in between {} into a string of text

```python
number = 6
# print("The number is" + number) --> error, cannot add string and int together like this
print("The number is", number) #careful, the , adds a space in between the datatypes
print(f"The number is {number}")
print("The number is " + str(number)) #using concatenation does not automatically add a space

print(number + number)
#print(str(number) + number) --> error
print(str(number) + str(number))
print(str(number + number))
#print("6" + number) --> careful, this works in c#, not in python
```

## Conditional Statements
if-structure
do keep in mind: zero is also an even number

```python
number = int(input("Give me a number: "))
if number > 0:
    print("BIG")
elif number < 0:
    print("SMALL")
else:
    print("EQUAL")


#check if a number is even or odd
if number % 2 == 0:
    print("even")
else:
    print("odd")
```

## Loops
for-loop / while-loop
- **for-loop** iterating over a sequence of numbers or a range, you know exactly how many times you need to loop
- **while-loop** looping as long as a condition is met, you don't know how many times you will be looping

```python
#keep in mind: a word is also a sequence, a sequence of characters

for c in "test":
    print(c)

for c in "test":
    print(c, end = " ") #adds space behind every char c, or whatever you put in the ""

for x in range(-5, 5):
    print(x) #5, so upper boundary not included!!!

for x in range(0, 11):
    print(f"5 * {x} = {x * 5}")
```

```python
word = input()

while word != "STOP":
    print("HELP")
    word = input() #without this line, the loop is infinite


x = 10
while x < 15:
    print(x + 2)
    x += 1 #needed to prevent infinite loop
    #x++  --> does not work!!!, neither does ++x
```

## Combining conditions
- and: both conditions have to be met
- or: at least one has to be met
- not: reverse of condition

```python
x,y,z = 1,2,3
print(x < y and y < z)
print(x < y or y < z)
print(x > y or y > z)
print(x > y and y > z)
print(x < y and y > z)
print(x < y or y > z)
print(not(x < y and y > z))
print(not(x < y and y > z))
```

### Exercise for loops:
FIZZBUZZ
Write a program that prints out the numbers from 1 -> 100 and think of the following
"Fizz" instead of the number if the number is divisible by 3 (only 3)
"Buzz" instead of the number if it is divisible by 5 (only 5)
"FizzBuzz" instead of the number if it is divisible by both 3 and 5!

```python
for i in range(1, 101):
    if i % 5 == 0 and i % 3 == 0:
        print("FizzBuzz", end=" ")
    elif i % 5 == 0:
        print("Buzz", end=" ")
    elif i % 3 == 0:
        print("Fizz", end=" ")
    else:
        print(i, end=" ")
```
