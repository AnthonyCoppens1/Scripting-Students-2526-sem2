## Input and output
Conversion types
- int -> integer (number)
- float -> decimal number
- str -> string of characters --> text!

```python
# find the sign of x
x = input()
print(x)
print(type(x))
# combining text and code in a line

print("Value of x:", x)

#converting between datatypes
int_x = int(x)
print("Value of x:", int_x) #add a comma to be able to work with multiple datatypes, whereas the + for concatenation
#will only work with a similar datatype

float_x = float(x)
print("Value of x:", float_x)

#we can also convert from anything to the original --> string
testnumber = str(float_x)
print("testnumber has the following datatype:", type(testnumber))
```

## Combining outputs --> concatenation
- '+' -> combining 2 strings or adding 2 numbers together, you have to provide a space if you want to split them
- ',' -> combining any datatype with each other, it provides a space for you to split them

```python
number = 5
print("The number is:", number)
print(number + number) #10
print(str(number) + str(number)) #55
print(str(number + number)) #10
# print(str(number) + number) --> results into an error
```

## Conditional statements

### Some exercises for if-statements
- Write an if-statement that checks if you are eligible to vote in Belgium. Based upon your actual age calculated on your birthyear. --> if a person is 18 or older, print --> "Eligible to vote" otherwise --> "NO VOTE FO JO" 
<br>USE DATETIME!!!

- Write an if-statement that checks if a given string is empty. --> either "EMPTY" "NOT EMPTY"


```python
#exercise 1
from datetime import datetime as dt

birthyear = int(input("What's your birthyear? "))
currentyear = dt.now().year
age = currentyear - birthyear

if age >= 18:
    print(f"{age} years old:")
    print("Eligible to vote")
else:
    print("NO VOTE FO YO")
```

```python
#exercise 2

S = input()
if S == "":
    print("EMPTY")
else:
    print("NOT EMPTY")

if len(S) == 0:
    print("EMPTY")
else:
    print("NOT EMPTY")


if not S:
    print("EMPTY")
else:
    print("NOT EMPTY")
```

## Loops
- <b>for-loop</b>: iterating over a sequence or a range, you know how many times you need to loop!
- **while-loop**: looping as long as the condition is true, you don't know how many times you need to loop, it 
depends on the condition

```python
#a word is also a sequence it's a sequence of characters
for c in "text":
    print(c)

for c in "text":
    print(c, end = " ") #not underneath each other, but behind with a space in between

#can replace "text" with a variable
s = "text"
for c in s:
    print(c)


#looping with numbers
for x in range(-5,5):
    print(x) #5 not included

for x in range(0,11):
    print(f"5 * {x} = {5*x}")
```

```python
#while loops
value = input("Enter something, stop with STOP")
while value != "STOP":
    print(value)
    value = input("Enter something, stop with STOP") #without this one --> infinite

x = 5
while x < 15:
    print(x, end = " ")
    x += 1
```

## Combining conditions
- **and**: both conditions have to be met
- **or**: at least one condition has to be met
- **not**: reverse of the condition

```python
x,y,z = 1,2,3
print(x < y and y < z)
print(x < y or y < z)
print(x > y or y > z)
print(x > y and y > z)
print(x < y and y > z)
print(x < y or y > z)
print(not(x < y and y > z))
print(not(x < y and y > z))
```
